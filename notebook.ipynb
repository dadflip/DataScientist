```python
# ── Cellule 0 : paquets de base ───────────────────────────────────────────────
import subprocess, sys
for pkg in ['pandas','numpy','matplotlib','seaborn','scikit-learn','scipy','ipywidgets']:
    subprocess.check_call([sys.executable,'-m','pip','install',pkg,'-q'],
                          stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
    print(f'  [OK] {pkg}')
print('Done.')
```

```python
# ── Initialisation ────────────────────────────────────────────────────────────
import sys, pathlib, traceback, urllib.request, os
_root = str(pathlib.Path().absolute())
if _root not in sys.path:
    sys.path.insert(0, _root)

import ipywidgets as widgets
from IPython.display import display
from ml_pipeline import load_config, PipelineState, styles

styles.apply_globals()
state = PipelineState()

# Charger config : local ou depuis GitHub
config_path = pathlib.Path("ml_pipeline/default.toml")
if not config_path.exists():
    config_path = pathlib.Path("config/default.toml")

if not config_path.exists():
    print("[INFO] Config not found locally, downloading from GitHub...")
    os.makedirs("ml_pipeline", exist_ok=True)
    github_url = "https://raw.githubusercontent.com/dadflip/DataScientist/main/ml_pipeline/default.toml"
    urllib.request.urlretrieve(github_url, "ml_pipeline/default.toml")
    config_path = pathlib.Path("ml_pipeline/default.toml")

state.config = load_config(str(config_path))
print(f'OK — config v{state.config.get("meta",{}).get("version","?")}')

def _show(obj):
    """Affiche obj.ui avec diagnostic d'erreur."""
    if not hasattr(obj, 'ui'):
        print('[ERROR] pas de .ui sur', type(obj).__name__)
        return
    display(obj.ui)
```
